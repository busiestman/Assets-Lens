```python
from datetime import timedelta, datetime
from pathlib import Path
import yfinance as yf


end_date = datetime(2026, 1, 1)
start_date = datetime(2016, 1, 1)

data = yf.download(
    ["EUNL.DE", "^GSPC"],
    start=start_date,
    end=end_date,
    auto_adjust=False,
    progress=False
)

df = data["Adj Close"].reset_index()

RAW_DATA_PATH = Path("../data/raw/stocks_raw.csv")

df.to_csv('../data/raw/stocks_raw.csv')
```
